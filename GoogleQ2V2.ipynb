```python
import numpy as np
import numexpr as ne
import math
import datetime

from functools import wraps
import errno
import os
import signal

'''
class TimeoutError(Exception):
    pass

def timeout(seconds=10, error_message=os.strerror(errno.ETIME)):
    def decorator(func):
        def _handle_timeout(signum, frame):
            raise TimeoutError(error_message)

        def wrapper(*args, **kwargs):
            signal.signal(signal.SIGALRM, _handle_timeout)
            signal.alarm(seconds)
            try:
                result = func(*args, **kwargs)
            finally:
                signal.alarm(0)
            return result

        return wraps(func)(wrapper)

    return decorator
'''

def isPrimeLargeSetQ3(n):
    a=5
    v=2
    if n % 2 == 0 or n % 3 == 0:
        return False
    if n == 2 or n == 3:
        return True
    c = 0
    while (a * a <= n) and (c < 1000):
        c += 1
        #print c
        if n % a == 0:
            #print c
            return False
        a=a+v
        v=6-v
    return True

def getDivisorLargeSetQ3(n):
    a=5
    v=2
    if n % 2 == 0:
        return 2
    if n % 3 == 0:
        return 3
    if n == 2 or n == 3:
        return 0
    c = 0
    while (a * a <= n) and (c < 1000):
        c += 1
        #print c
        if n % a == 0:
            return a
        a=a+v
        v=6-v
    return 0

def primeListCheck(valList):
    for i in valList:
        PrimeCheck = isPrimeLargeSetQ3(i)
        if PrimeCheck:
            print i
            return PrimeCheck
    return PrimeCheck

def getDivisor(NUM):
    for i in range(2,int(math.sqrt(NUM))+1):
        if NUM%i==0:
            return i

def baseN(candidate, base, DIG):
    candidateList = np.array([int(i) for i in candidate])
    powN = [base**x for x in (range(DIG))][::-1]
    return sum(candidateList*powN)

def isPrime(n):
    return all(n%i for i in xrange(2, n))

def isJamCoin(candidate):
    DIG = len(candidate)
    coinVal = [baseN(candidate,i,DIG) for i in range(2,11)]
    primeCheck = primeListCheck(coinVal)
    if primeCheck == False:
        return (True,candidate, coinVal)
    return (False,False)

def getCandidates(DIG,st,en):
    BINARY=np.array([2**i for i in range(DIG)])[::-1]

    MIN=sum(BINARY*np.array([int(imin) for imin in "1"+(DIG-1)*"0"]))
    MAX=sum(BINARY*np.array([int(imax) for imax in DIG*"1"]))   

    NUMLIST = MIN + np.arange(st,en)
    BINLIST = np.array(["{0:b}".format(i) for i in NUMLIST])
    BB = [[i[-1]=='1'] for i in BINLIST]
    BBI = np.array([i[0] for i in BB])

    CANDIDATES = BINLIST[BBI]
    return CANDIDATES

def getFinalCoins(DIG,NUMTOGET,st,en):    
    FINAL=[]
    GC = getCandidates(DIG,st,en)
    GCX = np.random.choice(GC,GC.shape[0],replace=False)
    #GCX=GC
    for i in GCX:
        J = isJamCoin(i)
        if J[0]==True:
            #tempFile = open("temp.txt",'a')
            FINAL.append(J)
            res= str(datetime.datetime.now()),"  Found Value, ",str(len(FINAL)), ", ", str(J)
            #print res
            #tempFile.write(res)
            #tempFile.close()
            if len(FINAL)==NUMTOGET:
                return FINAL

def final(DIG, NUMTOGET,st,en):
    Z = getFinalCoins(DIG, NUMTOGET,st,en)
    for (i,j,k) in Z:
        RESULT = [j,[getDivisorLargeSetQ3(v) for v in k]]
        #print str(RESULT[0]) + " " + " ".join(str(item) for item in RESULT[1])

def finalSave(inputFile,st,en):
    fp = open(inputFile)
    saveFile = "result_"+inputFile
    saveFile2 = "result_2"+inputFile
    
    tg = open(saveFile,'w')
    
    tg.write("Case #1:")
    tg.write("\n")
   
    for (ix, line) in enumerate(fp):
        if ix==0:
            pass
        else:
            val = line.split(" ")
            DIG=int(val[0])
            NUMTOGET=int(val[1])
    
    Z = getFinalCoins(DIG, NUMTOGET,st,en)
    for (i,j,k) in Z:
        tg2 = open(saveFile2,'a')
        RESULT = [j,[getDivisorLargeSetQ3(v) for v in k]]
        RESWRITE = str(RESULT[0]) + " " + " ".join(str(item) for item in RESULT[1])
        tg.write(RESWRITE)
        tg.write("\n")
        tg2.write(RESWRITE)
        tg2.write("\n")
        tg2.close()
    fp.close()
    tg.close()


```
